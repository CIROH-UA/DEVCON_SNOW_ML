```python
import pickle
import pandas as pd
RegionTrain= pickle.load(open(f"{training_path}/RegionTrain_SCA.pkl", "rb"))
```

```python
import pandas as pd
```

```python
Region_list = ['N_Sierras', 'S_Sierras_Low', 'S_Sierras_High']

for key in Region_list:
    RegionTrain[key].reset_index(inplace = True, drop = False)
del RegionTrain['Sierras_All']

combined_df = pd.concat(RegionTrain.values(), axis=0, ignore_index=True, sort=False)
combined_df.set_index('index', inplace=True)
```

```python
unique_values = combined_df['Date'].dt.strftime('%Y-%m-%d').unique()


# Convert to list (if you specifically need a list)
unique_date_list = list(unique_values)

datelist = unique_date_list
for date in datelist:
    for column in combined_df[combined_df['Date']==date].columns:
        median_value = combined_df[column].max()
        combined_df[column].fillna(value=median_value, inplace=True)
```

```python
RegionTrain['Sierras_All'] = combined_df

for key in Region_list:
    RegionTrain[key].set_index('index', inplace = True)
    
with open('RegionTrain_SCA_All.pkl', 'wb') as handle:
pickle.dump(RegionTrain, handle, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
import pickle

keystodel = ['N_Sierras_Snotel', 'S_Sierras_Snotel']
for date in datelist:
    path = f"Predictions/Hold_Out_Year/Predictions/Prediction_DF_SCA_{date}.pkl"
    
    # Load the file
    with open(path, "rb") as f:
        file = pickle.load(f)
    
    # Delete specified keys
    for key in keystodel:
        del file[key]
    
    # Concatenate the DataFrames
    combined_df = pd.concat(file.values(), axis=0, ignore_index=False, sort=False)
    
    # Add the combined DataFrame back to the dictionary
    file['Sierras_All'] = combined_df

    # Save the modified dictionary back to a pickle file
    with open(path, "wb") as f:  # You might want to save to a new file instead
        pickle.dump(file, f)
```
